```python
def get_bods(operator):
    operator_refs = [operator]
    params = SIRIVMParams(operator_refs=operator_refs)
    siri_response = bods.get_siri_vm_data_feed(params=params)
    siri = Siri.from_bytes(siri_response)
    #print(siri)
    vehicle_activities = siri.service_delivery.vehicle_monitoring_delivery.vehicle_activities
    
    df = pd.DataFrame([t.__dict__ for t in vehicle_activities ])
    df = df.drop('monitored_vehicle_journey', axis=1)
    
    df2 = pd.DataFrame([t.monitored_vehicle_journey.__dict__ for t in vehicle_activities ])
    df2 = df2.drop('vehicle_location', axis=1)
    
    df3 = pd.DataFrame([t.monitored_vehicle_journey.vehicle_location.__dict__ for t in vehicle_activities ])
    
    return pd.concat([df, df2, df3], axis=1)
```

```python
output = result_155.merge(result_prev_155,\
                          on=['vehicle_ref','direction_ref','vehicle_journey_ref','published_line_name'],\
                          how='left')
```

```python
output['delta_time_sec'] = (pd.to_datetime(output.recorded_at_time_x) -\
                    pd.to_datetime(output.recorded_at_time_y)).dt.total_seconds()
output['delta_time_sec'] = output['delta_time_sec'].fillna(0)
```

```python
import haversine as hs
def calc_distance(lon_x, lat_x, lon_y, lat_y):
    loc1=(lon_x,lat_x)
    loc2=(lon_y,lat_y)
    return hs.haversine(loc1,loc2)
```

```python
output['distance_km'] = \
    output.apply(lambda x: calc_distance(x.longitude_x,x.latitude_x,x.longitude_y,x.latitude_y), axis=1)
output['distance_km'] = output['distance_km'].fillna(0)
```

```python
output['avg_speed'] = output["distance_km"].divide((output["delta_time_sec"]/(60*60)), fill_value=0)
```

```python
output['avg_speed'] = output['avg_speed'].fillna(-1)
output[['vehicle_ref','delta_time_sec','distance_km','avg_speed']]
```

```python
result_155 = result_155.merge(output[['vehicle_ref','delta_time_sec','distance_km','avg_speed']],\
                      on=['vehicle_ref'], how='left')
```

```python
import urllib.request, json
import pandas as pd
from datetime import date
import random
from itertools import islice,cycle

def get_line_stops(line, direction):
    today = date.today()
    #print("Today's date:", today)

    df_line_stops = pd.DataFrame()

    url = "https://api.tfl.gov.uk/Line/"+line+"/Route/Sequence/"+direction+\
        "?serviceTypes=Regular&excludeCrowding=True"

    hdr ={
        # Request headers
        'Cache-Control': 'no-cache',
    }

    req = urllib.request.Request(url, headers=hdr)
    req.get_method = lambda: 'GET'
    response = urllib.request.urlopen(req)
    #print(response.getcode())
    #print(response.read())
    resp = json.loads(response.read())
        
    df_line_stops = pd.json_normalize(resp)
    #print(df_line_stops)
    line_stops = (df_line_stops.orderedLineRoutes.to_list())[0][0]['naptanIds']

    return line_stops
```

```python
import uuid
from faker import Faker
import datetime as dt
import random
import uuid

def assign_event(date, df_line_stops_coord):    
    #list_sentiments = ['negative','neutral','positive']
    list_event_types = ['Road work','Public demonstration','Sport event','Incident','Festival/Concert','Fair']
    #random_sentiment = random.choices(list_sentiments, weights=(10,40,50), k=1)
    weights_ = (20,20,20,20,20,20)
    if bool(len(pd.bdate_range(date, date))):
        weights_ = (30,10,10,30,10,10)
    else:
        weights_ = (0,20,20,30,20,10)
    
    random_event = random.choices(list_event_types, weights=weights_, k=1)
    #print(random_event)
    
    start_idx = random.choices(df_line_stops_coord.index.values.tolist())
    step = num1 = random.randint(1, 3)
    end_idx = 0
    if (start_idx[0]+step) < len(df_line_stops_coord.index.values.tolist()):
        end_idx = start_idx[0]+step
    else:
        end_idx = start_idx[0]
    #print(start_idx, step, end_idx)

    df_stops_event = df_line_stops_coord.iloc[range(start_idx[0],end_idx)].copy()
    df_stops_event['event_category'] = random_event[0]

    #print(df_stops_event[['stop_id','stop_lat','stop_lon','event_category']])
    
    return df_stops_event[['stop_id','stop_lat','stop_lon','event_category']]
```

```python
from randomtimestamp import randomtimestamp, random_date, random_time
from datetime import time

def event_time(report_date, event_category):    
    time_ = time(0, 0, 0)
    if event_category == 'Road work':
        time_ = random_time(start=time(7, 0, 0), end = time(17, 0, 0))
    elif event_category == 'Public demonstration':
        time_ = random_time(start=time(8, 0, 0), end = time(20, 0, 0))
    elif event_category == 'Sport event':
        time_ = random_time(start=time(9, 0, 0), end = time(23, 0, 0))
    elif event_category == 'Incident':
        time_ = random_time()
    elif event_category == 'Festival/Concert':
        time_ = random_time(start=time(18, 0, 0), end = time(23, 0, 0))
    elif event_category == 'Fair':
        time_ = random_time(start=time(7, 0, 0), end = time(17, 0, 0))
    
    mydatetime = dt.datetime.combine(report_date, time_)
    #print(mydatetime)
    return mydatetime
```

```python
def simulate_data(report_date, line, direction, stop_lat, stop_lon, event_category):

    report_datetime = event_time(report_date, event_category)
    #print(report_datetime)

    rand_lat = random.gauss(stop_lat, 0.001)
    rand_lon = random.gauss(stop_lon, 0.001)
    
    #fake = Faker()
    #start_date = datetime.date(year=2023, month=1, day=1)
    #event_date = fake.date_time_between(start_date=start_date, end_date='now')
    #limit_date = event_date+dt.timedelta(days=7)
    #report_date = fake.date_time_between(start_date=event_date, end_date=limit_date)
    user_id = uuid.uuid4()
    
    #print(user_id, ",", report_date, ",", line, ",", event_category,\
    #      ",", rand_lat, ",", rand_lon)
    
    return {"id":user_id, "report_date":report_date, "report_datetime": report_datetime,\
            "published_line_name":line, "event_category":event_category,\
            "rec_lat":rand_lat, "rec_lon":rand_lon,\
            "orig_stop_lat":stop_lat, "orig_stop_lon":stop_lon}
```

```python
def getRGBfromI(RGBint):
    RGBint = RGBint*RGBint*RGBint
    blue  =  RGBint & 255
    green = (RGBint >> 8) & 255
    red   =   (RGBint >> 16) & 255
    return red, green, blue
```

```python
def rgb_to_hex(rgb):
    return '#%02x%02x%02x' % rgb
```
