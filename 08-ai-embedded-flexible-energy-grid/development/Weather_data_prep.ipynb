```python
weather=weather.rename(columns={'Year':'year',
                                'Month':'month', 
                                'Day':'day-of-month',
                                'Time':'hour',
                                'Temp (C)':'temp',
                                'Rel Hum (%)':'hum',
                               } )
weather['hour']=weather['hour'].apply(lambda x: x.split(':')[0].lstrip('0') )
weather.loc[weather['hour']=='','hour']='0'
weather['hour']=weather['hour'].astype(int)
cols=['year','month','day-of-month','hour','temp','hum']
weather=weather[cols]
weather
```

### compute weather predictors

```python
weather_predictors=[]
feat=['temp','hum']

for f in feat:

    for h in range(1,25):
        var_name=f+'-h-minus-'+str(h)
        weather[var_name]=weather[f].shift(h)

    for i in [3,6,12,24]:
        var_name=f+'-h-'+str(i)
        weather_predictors.append(var_name)        
        var_group=[f+'-h-minus-'+str(h) for h in range(1,i+1)]
        weather[var_name]=weather[var_group].mean(axis=1)
 
weather
```

```python
weather_predictors
```

```python
for f in feat:
    for d in range(2,9):
        var_name=f+'-d-minus-'+str(d)
        weather_predictors.append(var_name)    
        weather[var_name]=weather[f].shift(d*24)
weather
```

```python
weather_predictors
```

```python
for f in feat:
    
    var_name1='avg_'+f+'-w1'
    weather_predictors.append(var_name1)
    var_group=[ c for c in weather.columns if f in c and '-d-minus-' in c]
    weather[var_name1]=weather[var_group].mean(axis=1)
    
    for w in range (2,5):
        var_name='avg_'+f+'-w'+str(w)
        weather_predictors.append(var_name)
        weather[var_name]=weather[var_name1].shift(w)

weather
```

```python
weather_predictors
```

```python
weather=weather[['year','month','day-of-month','hour']+weather_predictors] 
```

```python
weather_message=weather[weather_predictors].tail(1) 
weather_message.to_csv('weather_message.csv',index=False, header=False)
weather_message
```
